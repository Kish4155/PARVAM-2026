```python
import numpy as np

# Example of fast data crunching with NumPy
# Create a large array
data = np.random.rand(1000000)

# Perform fast vectorized operations
mean_val = np.mean(data)
std_val = np.std(data)
squared = np.square(data)

print(f"Mean: {mean_val}, Std: {std_val}, First 5 squared: {squared[:5]}")
```
