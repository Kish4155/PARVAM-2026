```python
import numpy as np

import matplotlib.pyplot as plt

# Demo data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# Line plot
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title('Line Plot')
plt.legend()

# Bar chart
plt.subplot(1, 2, 2)
plt.bar(categories, values)
plt.title('Bar Chart')

plt.tight_layout()
plt.show()
```
