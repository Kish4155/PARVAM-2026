```python
# Generate various data types and variables in Python

# Integer
age = 25
count = 100

# Float
height = 5.9
temperature = 98.6

# String
name = "Alice"
message = "Hello, World!"

# Boolean
is_active = True
is_completed = False

# List
fruits = ["apple", "banana", "cherry"]
numbers = [1, 2, 3, 4, 5]

# Tuple
coordinates = (10, 20)
rgb_color = (255, 128, 0)

# Dictionary
person = {
    "name": "Bob",
    "age": 30,
    "city": "New York"
}

# Set
unique_numbers = {1, 2, 3, 4, 5}
tags = {"python", "programming", "data"}

# None
result = None

# Print all variables
print(f"age: {age}, type: {type(age)}")
print(f"height: {height}, type: {type(height)}")
print(f"name: {name}, type: {type(name)}")
print(f"is_active: {is_active}, type: {type(is_active)}")
print(f"fruits: {fruits}, type: {type(fruits)}")
print(f"coordinates: {coordinates}, type: {type(coordinates)}")
print(f"person: {person}, type: {type(person)}")
print(f"unique_numbers: {unique_numbers}, type: {type(unique_numbers)}")
print(f"result: {result}, type: {type(result)}")
```

```python
if age >= 18:
    print("Age indicates an adult.")
else:
    print("Age indicates a minor.")

if is_active:
    print("The item is active.")
else:
    print("The item is not active.")
```

```python
if age >= 18:
    print("Age indicates an adult.")
    if is_active:
        print("The adult is active.")
    else:
        print("The adult is not active.")
else:
    print("Age indicates a minor.")
    if is_completed:
        print("The minor has completed the task.")
    else:
        print("The minor has not completed the task.")
```

```python
if age < 18:
    print("Minor")
elif age >= 18 and age < 65:
    print("Adult")
else:
    print("Senior")
```

```python
# Loop over the list of fruits
for fruit in fruits:
    print(f"I like {fruit}")

# Loop over numbers with index
for i, num in enumerate(numbers):
    print(f"Index {i}: {num}")

# Loop over dictionary items
for key, value in person.items():
    print(f"{key}: {value}")

# Loop over set
for tag in tags:
    print(f"Tag: {tag}")

# While loop example using count
while count > 0:
    print(f"Count: {count}")
    count -= 10
```
